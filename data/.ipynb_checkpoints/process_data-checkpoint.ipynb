Data loading and overview:

```python
df = pd.read_csv('imdb_raw.csv')
df.head()
```

Selecting the columns of interest:

```python
df = pd.DataFrame(df[['description', 'genre']])
df.head()
```

Checking and removing rows with null values:

```python
df.info()
```

```python
df.dropna(inplace=True)
df.info()
```

Identifying and removing duplicate values:

```python
df.nunique()
```

No duplicate lines were found, but removal would be through the following code:

```python
df.drop_duplicates(inplace=True)
df.nunique()
```

```python
df.head()
```

Converting the gender column into columns with values 0 and 1:

```python
df['genre'] = df['genre'].apply(lambda x: x.translate(str.maketrans({"[": "", "]": "", "'": '', " ": ''})))
df.head()
```

```python
dummies = df['genre'].str.get_dummies(sep=',')
dummies.head()
```

```python
if df.shape[1] <= 2:
    df = df.join(dummies)

df.head()
```

```python
df.drop('genre', axis=1, errors='ignore', inplace=True)
df.head()
```

Checking for movies without any classification by genre:

```python
df_aux = pd.DataFrame()
df_aux['ncategories'] = dummies.sum(axis=1)
df_aux.query('ncategories <= 0')
```

All movies are classified in the same category.

Computing statistics on the data:

```python
df_aux = pd.DataFrame(index=dummies.columns)
for column in dummies.columns:
    df_aux.at[column, 'count'] = len(dummies[dummies[column] == 1])
    df_aux.at[column, '%'] = df_aux.at[column, 'count'] / dummies.shape[0] * 100
df_aux['count'] = df_aux['count'].astype('int')
df_aux = df_aux.round(decimals=2)
df_aux.sort_values(by='count', ascending=False, inplace=True)
df_aux
```

Plotting visualizations:

```python
ax = df_aux['count'].plot(kind='bar', figsize=(15,5), title='Data Distribution by Genre', ylabel='Count');
for p in ax.patches:
    ax.annotate(p.get_height(), (p.get_x() * 1.000, p.get_height() * 1.000))
```

```python
ax = df_aux['%'].plot(kind='bar', figsize=(15,5), title='Data Distribution by Genre', ylabel='%');
for p in ax.patches:
    ax.annotate('%.2f'%(p.get_height()), (p.get_x() * 1.000, p.get_height() * 1.000))
```

The dataset presents a large difference in the number of samples for each movie genre. For example, out of 2334 samples, we only have one movie classified as Western.

The genre with the largest number of samples is Drama, which has almost twice as many samples (1065) as Comedy, with 553 samples.
